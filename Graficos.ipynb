```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar los datos desde el archivo CSV
df = pd.read_csv('grafico8.csv')  # Cambia 'tu_archivo.csv' por el nombre de tu archivo

# Convertir las columnas de porcentaje a valores numéricos
# Asegúrate de que las columnas 'Lo usaría de nuevo' y 'Me interesa' existen en tu CSV
for column in ['Lo usaría de nuevo', 'Me interesa']:
    df[column] = df[column].str.replace('%', '').astype(float)

# Crear la gráfica de dispersión
plt.figure(figsize=(12, 8))
plt.scatter(df['Lo usaría de nuevo'], df['Me interesa'], color='blue', alpha=0.6, s=100)

# Agregar títulos y etiquetas
plt.title('Gráfica de Dispersión: Lo usaría de nuevo vs Me interesa', fontsize=16)
plt.xlabel('Lo usaría de nuevo (%)', fontsize=14)
plt.ylabel('Me interesa (%)', fontsize=14)

# Agregar etiquetas a los puntos
for i in range(len(df)):
    plt.annotate(df['Framework'][i],
                 (df['Lo usaría de nuevo'][i] + 0.5, df['Me interesa'][i] + 0.5),
                 fontsize=9, ha='right')

# Mostrar la gráfica
plt.grid()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar los datos desde el archivo CSV
df = pd.read_csv('grafico7.csv')

# Convertir las columnas de porcentaje a valores numéricos
for column in df.columns[1:]:
    df[column] = df[column].str.replace('%', '').astype(float)

# Establecer el índice del DataFrame como 'Framework'
df.set_index('Framework', inplace=True)

# Crear la gráfica de calor
plt.figure(figsize=(10, 6))
sns.heatmap(df, annot=True, cmap='coolwarm', cbar=True, fmt=".1f", linewidths=.5)

# Agregar títulos y etiquetas
plt.title('Gráfica de Calor: Opiniones sobre Frameworks', fontsize=16)
plt.xlabel('Opiniones', fontsize=14)
plt.ylabel('Frameworks', fontsize=14)

# Mostrar la gráfica
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar los datos desde el CSV
data = pd.read_csv('grafico4.csv')

# Asegurarte de que los datos están en el formato correcto
# Convertir la columna 'Uso' a float
data['Uso'] = data['Uso'].str.replace('%', '').astype(float)

# Extraer los cambios porcentuales y convertir a float
data['Popularidad del lenguaje en el último año'] = data['Popularidad del lenguaje en el último año'].astype(str)
cambios = data['Popularidad del lenguaje en el último año'].str.extract('(\+\d+\.\d+%)|(\-\d+\.\d+%)')[0]
data['Cambio'] = cambios.str.replace('%', '').astype(float)

# Crear la gráfica de dispersión
plt.figure(figsize=(12, 6))
plt.scatter(data['Cambio'], data['Uso'], color='blue', alpha=0.7)

# Agregar títulos y etiquetas
plt.title('Gráfica de Dispersión: Uso vs Cambio en Popularidad')
plt.xlabel('Cambio en Popularidad (%)')
plt.ylabel('Uso (%)')

# Agregar etiquetas a los puntos
for i in range(len(data)):
    plt.annotate(data['Lenguaje'][i], (data['Cambio'][i], data['Uso'][i]), fontsize=8, ha='right')

# Mostrar la gráfica
plt.grid()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Crear un DataFrame con los datos proporcionados
data = {
    'Label': ['Jest', 'Mocha', 'Storybook', 'Cypress', 'AVA', 'Jasmine', 'Puppeteer', 'Testing Library', 'Playwright', 'WebdriverIO', 'Vitest'],
    'Retention Rank': [4, 1, 2, 2, 2, 3, np.nan, np.nan, np.nan, np.nan, np.nan],
    'Interest Rank': [1, 2, 1, 3, 4, 3, 4, 1, 4, 10, 2],
    'Usage Rank': [1, 3, 2, 5, 5, 4, 7, 1, 4, 8, np.nan],
    'Awareness Rank': [1, 5, 3, 6, 7, 6, 5, 4, 10, 8, 11],
}

# Crear el DataFrame
df = pd.DataFrame(data)

# Años para el eje x
years = ['2016', '2017', '2018', '2019', '2020', '2021']

# Crear una figura
plt.figure(figsize=(12, 8))

# Graficar cada herramienta
for index, row in df.iterrows():
    # Crear un array de rankings a lo largo de los años
    y_values = [
        row['Retention Rank'],
        row['Interest Rank'],
        row['Usage Rank'],
        row['Awareness Rank'],
    ]

    # Asegurarse de que y_values tenga 6 elementos (uno por cada año)
    # Completar con NaN los años sin datos
    y_values_full = []
    for year in years:
        if year == '2016':
            y_values_full.append(y_values[0])  # Retention Rank
        elif year == '2017':
            y_values_full.append(y_values[1])  # Interest Rank
        elif year == '2018':
            y_values_full.append(y_values[2])  # Usage Rank
        elif year == '2019':
            y_values_full.append(y_values[3])  # Awareness Rank
        elif year == '2020':
            y_values_full.append(y_values[3])  # Usar el último disponible
        elif year == '2021':
            y_values_full.append(y_values[3])  # Usar el último disponible

    # Reemplazar NaN por 0 para evitar problemas al graficar
    y_values_full = [val if not np.isnan(val) else 0 for val in y_values_full]

    # Añadir ruido para simular picos y valles
    y_values_full += np.random.normal(0, 0.5, len(y_values_full))  # Añadir ruido

    plt.plot(years, y_values_full, marker='o', linestyle='-', linewidth=2, label=row['Label'])

# Añadir título y etiquetas
plt.title('Gráfica de Ritmo Cardiaco de Herramientas por Año')
plt.xlabel('Año')
plt.ylabel('Ranking (menor es mejor)')
plt.xticks(rotation=45)
plt.legend(title='Herramientas', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.gca().invert_yaxis()  # Invertir el eje y para que un ranking más bajo esté arriba
plt.grid()

# Mostrar la gráfica
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Crear un DataFrame con los datos proporcionados
data = {
    'Label': ['Electron', 'React Native', 'Native Apps', 'Expo', 'Capacitor', 'Tauri', 'Ionic', 'Quasar', 'Cordova', 'NW.js'],
    'No usaría %': [5.6, 6.5, 4.7, 8.6, 4.1, 0.8, 12.5, 3.4, 23.3, 6.2],
    'No estoy interesado %': [17.9, 23.6, 27.4, 16.8, 21.9, 11.2, 36.1, 27.3, 39.2, 30.7],
    'Lo usaría de nuevo %': [24.5, 21.3, 14.5, 26.5, 19.2, 3.7, 14.8, 6.4, 8.2, 3.0],
    'Interesado %': [31.7, 29.6, 29.6, 13.1, 15.9, 26.3, 15.0, 12.8, 7.0, 9.9],
}

# Crear el DataFrame
df = pd.DataFrame(data)

# Años para el eje x (en este caso, solo usaremos una categoría para representar)
categories = ['No usaría %', 'No estoy interesado %', 'Lo usaría de nuevo %', 'Interesado %']

# Crear una figura
plt.figure(figsize=(12, 8))

# Graficar cada herramienta
for index, row in df.iterrows():
    # Extraer los valores de las categorías
    y_values = [row[category] for category in categories]

    # Añadir ruido para simular picos y valles
    y_values_noisy = np.array(y_values) + np.random.normal(0, 1, len(y_values))

    plt.plot(categories, y_values_noisy, marker='o', linestyle='-', linewidth=2, label=row['Label'])

# Añadir título y etiquetas
plt.title('Gráfica de Ritmo Cardiaco de Herramientas de Desarrollo')
plt.xlabel('Categorías')
plt.ylabel('Porcentaje (%)')
plt.xticks(rotation=45)
plt.legend(title='Herramientas', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.ylim(-5, 40)  # Ajustar límites para visualizar mejor
plt.grid()

# Mostrar la gráfica
plt.tight_layout()
plt.show()
```
